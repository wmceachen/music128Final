**Global Plastics Production**


In 1950 the world produced only 2.2 million tons per year. Since then, annual production has increased nearly 200-fold, reaching 419 million tons in 2015. For context, this is **<font color = ##008700>roughly equivalent to the mass of two-thirds of the world population</font>**.


The short downturn in annual production in 2009 and 2010 was predominantly the result of the 2008 global financial crisis â€” a similar dent is seen across several metrics of resource production and consumption, including energy.


```python
ww_df = pd.read_csv('global-plastics-production.csv')
x = ww_df.Year
y = ww_df.Global
plt.plot(x, y)
plt.title("Global Plastics Production 1950 to 2015")
plt.xlabel("Year", fontsize=18)
plt.ylabel("Global Plastic Production (million tons)", fontsize=14)
```

How many new plastic bags & straws do you use a week? (This may include bags for shopping, take-out food and online shopping).

```python
wk_bags = wg.IntSlider(description='Plastic Bags:')
wk_straws = wg.IntSlider(description='Plastic Straws:')
age_y = wg.IntSlider(description="Age")
display(wk_bags, wk_straws, age_y)
```

```python
print("Monthly, your single-use plastic habits for these items use...")
print("Energy in MJ: " + str(wk_bags.value *0.48 *3.5 + wk_straws.value * 0.3 *3.5))
print("CO2 emissions in pounds: " + str((wk_bags.value*3.5 + wk_straws.value*3.5)/3600))
print("This is equivalent (in terms of CO2 emissions) to driving " + str((wk_bags.value *0.48*3.5 + wk_straws.value * 0.3*3.5)/6.7) + " miles in a car.")

```

On average (in the US), single-use plastics like bags and straws are used for 12 minutes before they are discarded (and can take over 500 years to degrade). The following cell will calculate the number of single-use plastics disposed via recycling, incinerated, and discarded, according to these national averages and your identified age and consumption habits.

| Disposal | % of Single use Plastics  |
|---------|---------------|
| Recycled | 9% |
| Incinerated | 12% |
| Discarded | 79% |


```python

data = [["Recycled", ((wk_bags.value + wk_straws.value)* 52.1429 * age_y.value * 0.09)], ["Incinerated", ((wk_bags.value + wk_straws.value)* 52.1429 * age_y.value * 0.12)],
       ["Discarded", ((wk_bags.value + wk_straws.value)* 52.1429 * age_y.value * 0.79)]]
pd.DataFrame(data, columns=["Disposal", "Plastics Disposed in Lifetime"])
```

See the following table to visualize single-use plastic usage if all UC Berkeley students, California Students, and US Students followed your same consumption habits. In the rightmost column, notice the difference in CO2 emissions if each person reduced their weekly plastics use by 3 straws or 3 bags.

The impact of a successful campaign globally against the use of single-use plastics has a larger dent in oil demand than the dent caused by electric cars. In other words, effective regulations on plastic could reduce oil demand by at least as much as the adoption of electric cars 20 years from now.

```python
ucb_students = 41910
us_all_college_stu = 19912960
us_students = 81065400
avg_adult_friends = 40
```

```python
def yourPlot(s, b):
    c = [1, avg_adult_friends, ucb_students, us_all_college_stu, us_students]
    d = []
    for i in c:
        d.append(i*(b *0.48 +  s* 0.3)/6.7)
    return [c,d]
```

```python
b = yourPlot(wk_straws.value, wk_bags.value)[1]
g = yourPlot(max(0, (wk_straws.value - 3)), (max(0, (wk_bags.value - 3))))[1]
dif = [b[0]/100 - g[0]/100, b[1]/100 - g[1]/100, b[2]/100 - g[2]/100, b[3]/100 - g[3]/100, b[4]/100 -g[4]/100]
data = [["Personal", b[0]/100, g[0]/100, dif[0]], ["Friends", b[1]/100, g[1]/100, dif[1]], ["UCB Students", b[2]/100, g[2]/100, dif[2]],["US College Students", b[3]/100, g[3]/100, dif[3]],["US Students", b[4]/100, g[4]/100, dif[4]]]
pd.DataFrame(data, columns=["Group", "Emissions (100's of Driven Miles)", "Reduced Plastics Use by 3", "Difference (100's of Miles)"])
```
